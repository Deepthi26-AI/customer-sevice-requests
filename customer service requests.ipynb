```python
Complaints=ds.dropna(subset=["Complaint Type"])
Complaints=ds.groupby("Complaint Type")
sortedType = Complaints.size().sort_values(ascending = False)
sortedType = sortedType.to_frame('count').reset_index()
sortedType
sortedType.head(15)
```

```python
sortedType = sortedType.head()
plt.figure(figsize=(6,6))
plt.pie(sortedType['count'],labels=sortedType["Complaint Type"], autopct="%1.1f%%")
plt.show()
```
